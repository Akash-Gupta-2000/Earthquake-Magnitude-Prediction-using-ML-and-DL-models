```python
import matplotlib.pyplot as plt
import pandas as panda
```

## Read input data ##

```python
inputData = panda.read_csv('predictions.csv')
inputData
```

## Plots ##

## Algorithms' Precision for Each Radius ##

```python
plt.figure(figsize=(12,6))
plt.title('Algorithm Precision for each radius')
plt.xlabel('Radius')
plt.ylabel('Magnitude')
plt.plot( 'Radius', 'RF', data = inputData, marker='o', markerfacecolor='blue', markersize=7, color='skyblue', linewidth=2)
plt.plot( 'Radius', 'MLP', data = inputData, marker='o', markerfacecolor='orange', markersize=7, color='red', linewidth=2)
plt.plot( 'Radius', 'SVR', data = inputData, marker='o', markerfacecolor='green', markersize=7, color='green', linewidth=2)
plt.plot( 'Radius', 'Real_Magnitude', data = inputData, linestyle='dashed', color='green', linewidth=2)
plt.legend()
```

```python
plt.title('Random Forest Classifier')
plt.xlabel('Radius')
plt.ylabel('Magnitude')
plt.plot( 'Radius', 'RF', data = inputData, marker='o', markerfacecolor='blue', markersize=7, color='skyblue', linewidth=2)
plt.plot( 'Radius', 'Real_Magnitude', data = inputData, linestyle='dashed', color='green', linewidth=2)
plt.legend()
```

```python
plt.title('Multilayer Perceptron')
plt.xlabel('Radius')
plt.ylabel('Magnitude')
plt.plot( 'Radius', 'MLP', data = inputData, marker='o', markerfacecolor='orange', markersize=7, color='red', linewidth=2)
plt.plot( 'Radius', 'Real_Magnitude', data = inputData, linestyle='dashed', color='green', linewidth=2)
plt.legend()
```

```python
plt.title('Support Vector Machines')
plt.xlabel('Radius')
plt.ylabel('Magnitude')
plt.plot( 'Radius', 'SVR', data = inputData, marker='o', markerfacecolor='blue', markersize=7, color='purple', linewidth=2)
plt.plot( 'Radius', 'Real_Magnitude', data = inputData, linestyle='dashed', color='green', linewidth=2)
plt.legend()

# plt.savefig('magVSradSVM.png')
```
