```python
import matplotlib.pyplot as plt
import pandas as panda
```

```python
inputData = panda.read_csv('predictions.csv')
inputData
```

## Algorithms' Precision

```python
plt.figure(figsize=(12,6))
plt.title('Algorithm Precision for each radius')
plt.xlabel('Radius')
plt.ylabel('Magnitude')
plt.plot( 'Radius', 'RF', data = inputData, marker='o', markerfacecolor='skyblue', markersize=7, color='skyblue', linewidth=2)
plt.plot( 'Radius', 'MLP', data = inputData, marker='o', markerfacecolor='red', markersize=7, color='red', linewidth=2)
plt.plot( 'Radius', 'SVR', data = inputData, marker='o', markerfacecolor='blue', markersize=7, color='blue', linewidth=2)
plt.plot( 'Radius', 'Real_Magnitude', data = inputData, linestyle='dashed', color='green', linewidth=2)
plt.legend()
```

## Random Forest

```python
plt.title('Random Forest')
plt.xlabel('Radius')
plt.ylabel('Magnitude')
plt.plot( 'Radius', 'RF', data = inputData, marker='o', markerfacecolor='skyblue', markersize=7, color='skyblue', linewidth=2)
plt.plot( 'Radius', 'Real_Magnitude', data = inputData, linestyle='dashed', color='green', linewidth=2)
plt.legend()
```

## Multilayer Perceptron

```python
plt.title('Multilayer Perceptron')
plt.xlabel('Radius')
plt.ylabel('Magnitude')
plt.plot( 'Radius', 'MLP', data = inputData, marker='o', markerfacecolor='red', markersize=7, color='red', linewidth=2)
plt.plot( 'Radius', 'Real_Magnitude', data = inputData, linestyle='dashed', color='green', linewidth=2)
plt.legend()
```

## Support Vector Machines

```python
plt.title('Support Vector Machines')
plt.xlabel('Radius')
plt.ylabel('Magnitude')
plt.plot( 'Radius', 'SVR', data = inputData, marker='o', markerfacecolor='blue', markersize=7, color='blue', linewidth=2)
plt.plot( 'Radius', 'Real_Magnitude', data = inputData, linestyle='dashed', color='green', linewidth=2)
plt.legend()
```
